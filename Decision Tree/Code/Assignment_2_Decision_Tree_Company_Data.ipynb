```python
def split(num):
    if num < 5:
        return "low sale"
    elif num>= 5 and num<10:
        return "average sale"
    else:
        return "high sale"

company.Sales=company.Sales.apply(split)
```

```python
company
```

```python
corr=company.corr()
sns.heatmap(corr,annot=True)

```

```python
le=LabelEncoder()
```

```python
company["ShelveLoc"]=le.fit_transform(company["ShelveLoc"])
company["Urban"]=le.fit_transform(company["Urban"])
company["US"]=le.fit_transform(company["US"])
```

```python
y=company.iloc[:,0]
x=company.iloc[:,1:11]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model=DE(criterion="entropy",max_depth=5)
model.fit(x_train,y_train)
```

```python
from sklearn import tree
```

```python
tree.plot_tree(model);
```

```python
fn=["Sales","CompPrice","Income","Advertising","Population","Price","ShelveLoc","Age","Education","Urban","US"]
cn=["low sale","average sale","high sale"]
fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(1.5,1.5),dpi=5000)
tree.plot_tree(model,feature_names=fn,class_names=cn,filled=True);
```

```python
preds=model.predict(x_test)
```

```python
preds
```

```python
confusion_matrix(y_test,preds)
```

```python
accuracy_score(y_test,preds)
```
